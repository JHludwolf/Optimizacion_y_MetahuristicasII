```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Evolutionary Programming/datos.csv')
X = df['x'].values
Y = df['y'].values

plt.figure()
plt.plot(X,Y, '.')
plt.show()
```

```python
bounds = [   [-100, 100],    # Rango de A
            [-100, 100],    # Rango de B
            [-100, 100] ]   # Rango de C
```

```python
def objective_function(P, X, Y):
    A, B, C = P[0], P[1], P[2]
    Yp = A*np.sin(B*X + C*np.pi)
    return np.sum(np.abs(Y-Yp))
```

```python
def draw_solution(P, X, Y):
    A, B, C = P[0], P[1], P[2]
    Yp = A*np.sin(B*X + C*np.pi)

    plt.figure()
    plt.plot(X, Y, '.')
    plt.plot(X, Yp)
    plt.show()
```

```python
from differential_evolution import differential_evolution
```

```python
chrom, score = differential_evolution(objective_function, bounds, args=(X,Y), niter=1000)
```

```python
draw_solution(chrom, X, Y)
```
